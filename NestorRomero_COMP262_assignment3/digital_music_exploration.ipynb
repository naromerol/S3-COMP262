```python
import numpy as np
import pandas as pd
import os, gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

base_path = os.getcwd()
data = getDF(os.path.join(base_path,'meta_Digital_Music.json.gz'))
```
